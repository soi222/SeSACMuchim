```python
# YOLOv8 ëª¨ë¸ ë¡œë“œ
model = YOLO("yolov8n.yaml")  # build a new model from scratch
model = YOLO("yolov8n.pt")    # load a pretrained model (recommended for training)
```

```python
# ëª¨ë¸ í•™ìŠµ
results = model.train(
    data= "/content/Drive/MyDrive/SeSAC_Final_Project_á„‰á…¢á„Šá…¡á†¨á„†á…®á„á…µá†·/model/yolov8/yolov8_pothole.yaml",  # ìœ„ì—ì„œ í™•ì¸í•œ yaml ê²½ë¡œ ì‚¬ìš©
    epochs=100,
    imgsz=640,
    batch=16,
    name='yolov8n_custom',
    device=0  # GPU ì‚¬ìš©ì‹œ 0, CPU ì‚¬ìš©ì‹œ 'cpu'
)
```
