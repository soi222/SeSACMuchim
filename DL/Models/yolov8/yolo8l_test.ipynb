```python
# YOLOv8 Î™®Îç∏ Î°úÎìú
model = YOLO("yolov8l.yaml")  # build a new model from scratch
model = YOLO("yolov8l.pt")    # load a pretrained model (recommended for training)
```

```python
# Î™®Îç∏ ÌïôÏäµ
results = model.train(
    data= "/content/Drive/MyDrive/SeSAC_Final_Project_·Ñâ·Ö¢·Ñä·Ö°·Ü®·ÑÜ·ÖÆ·Ñé·Öµ·Ü∑/model/yolov8/yolov8_pothole.yaml",  # ÏúÑÏóêÏÑú ÌôïÏù∏Ìïú yaml Í≤ΩÎ°ú ÏÇ¨Ïö©
    epochs=100,
    imgsz=640,
    batch=16,
    name='yolov8l_custom',
    device='cpu'  # GPU ÏÇ¨Ïö©Ïãú 0, CPU ÏÇ¨Ïö©Ïãú 'cpu'
)
```

```python
results = model.val()  # evaluate model performance on the validation set
```

```python
# ÏÉòÌîå Ïù¥ÎØ∏ÏßÄÎ°ú ÏòàÏ∏° ÌÖåÏä§Ìä∏
results = model("/content/Drive/MyDrive/SeSAC_Final_Project_·Ñâ·Ö¢·Ñä·Ö°·Ü®·ÑÜ·ÖÆ·Ñé·Öµ·Ü∑/data/TEST_dataset/Pothole/3_pothole_total_619/Pothole_Image_Data/1.jpg") # predict on an image
```

```python
from PIL import Image
# Í≤∞Í≥º ÏãúÍ∞ÅÌôî
for r in results:
    im_array = r.plot()
    im = Image.fromarray(im_array[..., ::-1])
    display(im)
```
